```python
import pandas as pd
import numpy as np

# 加载数据集
data_path = 'ratings.csv'
df = pd.read_csv("ratings.csv")

df
```

```python
# 定义Bandwagon Attack的参数
num_fake_users = 10  # 虚假用户的数量
num_items_to_rate = 5  # 每个虚假用户评分的物品数量
fake_rating = 5.0  # 假设数据集中最高评分

max_userId = max(df.userId)
new_df = df

# 创建一个空的数据框来存储虚假用户数据
fake_users_df = pd.DataFrame(columns=new_df.columns)

# 添加虚假用户
for i in range(num_fake_users):
    for item in range(num_items_to_rate):
        new_row = {
            'userId': max_userId + i + 1,
            'movieId': np.random.choice(new_df.movieId),    # 随机选择电影
            'rating': fake_rating,                          # 目标评分
            'timestamp':np.random.choice(new_df.timestamp)  # 随机选择时间
        }
    
        # 将字典转换为DataFrame
        new_row_df = pd.DataFrame([new_row])

        # 使用pd.concat将新的DataFrame行追加到现有DataFrame中
        new_df = pd.concat([new_df, new_row_df], ignore_index=True)

# 将投毒后的数据集保存为新的CSV文件
poisoned_data_path = 'poisoned_ratings.csv'
new_df.to_csv(poisoned_data_path, index=False)

print(f'投毒后的数据集已保存到 {poisoned_data_path}')
```
